Here is a simple example of a camera. Note, you cannot run video in Jupyter, but you can grab one image from an OpenCV camera.

```python
#!/usr/bin/env python

import numpy as np
import cv2
from matplotlib import pyplot as plt
import time

cam = cv2.VideoCapture(0)

if not cam.isOpened():
    print("... ops, failed to open camera")
    exit(1)

try:
    while True:
        ok, frame = cam.read()
        if ok:
            # cv2.imshow('movie', frame)
            k = cv2.waitKey(1)
            if k == ord('q'):
                cam.release()            # close camera
                cv2.destroyAllWindows()  # clean up GUI
                exit()
        time.sleep(0.033)
except KeyboardInterrupt:
    pass
        
cam.release()
```

````python
import numpy as np
import cv2
import time
import platform
from opencv_camera import bgr2rgb
from matplotlib import pyplot as plt

from ipywidgets import Video
````

````python
movie = cv2.VideoCapture("megamind.mp4")

# the first frame is black, grab a later frame
for _ in range(50): 
    ok, im = movie.read()

im = bgr2rgb(im)
plt.subplots(1,1,figsize=(9,9))
plt.imshow(im)
plt.title(f"{im.shape}");
````

````python
frames = []

cam = cv2.VideoCapture(0)

if not cam.isOpened():
    print("... ops, failed to open camera")
    raise Exception

for _ in range(10):
    ok, frame = cam.read()
    if ok:
        frames.append(frame)
        time.sleep(1)
        
cam.release()
````
